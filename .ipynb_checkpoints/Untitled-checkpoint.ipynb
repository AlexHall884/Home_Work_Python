Дана функция f(x) = (sin(x)/2 + cos(x)/2)^2 * sin(x)

1. Определить корни

```python
from sympy import*
y = trigsimp(sin(x)/2 + cos(x)/2)**2 * sin(x)
y
```

```python
print(f'Корни функции {(sin(x)/2 + cos(x)/2)**2 * sin(x)} :')
solveset(y,x)
```

2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция Убывает


Находим экстремумы

```python
y = (sin(x)/2 + cos(x)/2)**2 * sin(x)
list_extremum = solve(y, x)
list_extremum = sorted(extremum)
print(f'Точки экстремумов -> {list_extremum}')
```

Находим периоды

```python
t = pi
while True:
    a_1 = y.subs(x, 0)
    a_2 = y.subs(x, 0+t)
    print(a_1)
    if a_1 == a_2:
        break
    else:
        
        t+=pi        
        
point = list_extremum[0] + t
list_diff_1 = list_extremum[:]
list_diff_1.append(point)
list_diff_1 = sorted(list_diff_1)
for i in range(1, len(list_diff_1)):
    a_1 = y.subs(x,list_diff_1[i-1] + 0.000001)
    a_2 = y.subs(x,list_diff_1[i] - 0.000001)
    if a_1 < a_2:
        print(f'{i}. Функция возрастает на промежутке {list_diff_1[i-1]} : {list_diff_1[i]}.')
    else:
        print(f'{i}. Функция убывает на промежутке {list_diff_1[i-1]} : {list_diff_1[i]}.')
```

4.	Построить график

```python
graphic = plot(y, title = 'График f(x) = (sin(x)/2 + cos(x)/2)^2 * sin(x) ',line_color = 'blue')
```

```python
graphic_period = plot(y,(x, 0, 2*pi), title = 'График периода f(x) = (sin(x)/2 + cos(x)/2)^2 * sin(x) ',line_color='green')
```

5. Вычислить вершину

```python
period=0
for diff in list_extremum:
        point = y.subs(x, diff)
        period+=1
        print(f'Вершина функции в {period} периоде: x = {diff}, y = {point}')
```

6. Определить промежутки, на котором f > 0

```python
print('Промежутки на которых функция больше нуля: ')
solve_univariate_inequality(y > 0, x, relational = False ,domain = Interval(0, 2*pi))
```

7. Определить промежутки, на котором f < 0

```python
print('Промежутки на которых функция меньше нуля: ')
solve_univariate_inequality(y < 0, x, relational = False ,domain = Interval(0, 2*pi))
```
